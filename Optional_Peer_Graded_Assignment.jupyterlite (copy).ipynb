```python

State the null and alternative hypotheses.
Perform the relevant statistical test.
Report the test statistics and conclusion based on the p-value.
1. T-Test: Difference in Median Value of Houses Bounded by the Charles River or Not (CHAS)
Hypotheses:
Null Hypothesis (H₀): There is no significant difference in the median value of houses (MEDV) between houses near the Charles River (CHAS = 1) and those that are not (CHAS = 0).
Alternative Hypothesis (Hₐ): There is a significant difference in the median value of houses between houses near the Charles River (CHAS = 1) and those that are not (CHAS = 0).
```

```python
from scipy.stats import ttest_ind

# Subset data based on CHAS values
chas_1 = boston_df[boston_df["CHAS"] == 1]["MEDV"]
chas_0 = boston_df[boston_df["CHAS"] == 0]["MEDV"]

# Perform T-test
t_stat, p_value_ttest = ttest_ind(chas_1, chas_0)

# Output the results
t_stat, p_value_ttest
```

```python
Conclusion:
If the p-value ≥ 0.05, we fail to reject the null hypothesis and conclude there is no significant difference.
```

```python
Hypotheses:
Null Hypothesis (H₀): There is no significant difference in the median value of houses (MEDV) across the three age groups of properties (discretized AGE variable). That is, the mean MEDV is the same for all age groups.

Alternative Hypothesis (Hₐ): There is a significant difference in the median value of houses (MEDV) across the three age groups of properties. That is, at least one age group has a different mean MEDV.
```

```python
import pandas as pd
from scipy.stats import f_oneway

# Create the age group labels
bins = [0, 35, 70, 100]
labels = ["35 years and younger", "Between 35 and 70 years", "70 years and older"]
boston_df["AGE_group"] = pd.cut(boston_df["AGE"], bins=bins, labels=labels)

# Perform ANOVA test
anova_groups = [boston_df[boston_df["AGE_group"] == label]["MEDV"] for label in labels]
f_stat, p_value_anova = f_oneway(*anova_groups)

# Output the results
f_stat, p_value_anova
```

```python
Conclusion:
Fail to reject the null hypothesis (H₀) and conclude that there is no significant difference in the mean MEDV across the three age groups.
```

```python
Null Hypothesis (H₀): There is no significant linear relationship between nitric oxide concentrations (NOX) and the proportion of non-retail business acres (INDUS). This means the Pearson correlation coefficient between NOX and INDUS is zero.

Alternative Hypothesis (Hₐ): There is a significant linear relationship between nitric oxide concentrations (NOX) and the proportion of non-retail business acres (INDUS). This means the Pearson correlation coefficient is not zero.
```

```python
from scipy.stats import pearsonr

# Perform Pearson correlation between NOX and INDUS
corr_coeff, p_value_corr = pearsonr(boston_df["NOX"], boston_df["INDUS"])

# Output the results
corr_coeff, p_value_corr
```

```python
Conclusion
 p-value ≥ 0.05: Fail to reject the null hypothesis (H₀) and conclude that there is no significant linear relationship between NOX and INDUS.
```

```python
Null Hypothesis (H₀): There is no significant relationship between the weighted distance to the five Boston employment centres (DIS) and the median value of homes (MEDV). In other words, an additional unit increase in the weighted distance does not affect the median value of homes.

Alternative Hypothesis (Hₐ): There is a significant relationship between the weighted distance to the five Boston employment centres (DIS) and the median value of homes (MEDV). In other words, an additional unit increase in the weighted distance significantly affects the median value of homes.
```

```python
import statsmodels.api as sm

# Define the independent variable (DIS) and the dependent variable (MEDV)
X = sm.add_constant(boston_df["DIS"])  # Add constant term for the intercept
y = boston_df["MEDV"]

# Perform the regression analysis
model = sm.OLS(y, X).fit()

# Get the summary of the regression model
regression_summary = model.summary()

# Output the summary of regression results
regression_summary
```

```python
conclusion
 we reject the null hypothesis and conclude that there is a significant impact of DIS on MEDV.
```
